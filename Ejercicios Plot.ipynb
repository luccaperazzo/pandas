```python
#6
import pandas as pd
df = pd.read_csv(r'C:\Users\Usuario\Desktop\csv\alphabet.csv')
df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(10, 6))
start_date = pd.to_datetime('2020-04-09')
end_date = pd.to_datetime('2020-05-14')
# Filter the DataFrame to include data only between the two dates
filtered = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
plt.barh(filtered['Date'], filtered['Open'], color="red", label='Open')
plt.barh(filtered['Date'], filtered['Close'], left=filtered['Open'], color="g", label='Close')
plt.xlabel('Price')
plt.ylabel('Date')
plt.title('Alphabet Inc. (Google) Stock Prices')
plt.tight_layout()
plt.show()
```
